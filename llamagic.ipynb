```python
from datasets import Dataset
import math

# Function to generate the "input" string by concatenating specified fields
def create_input(example, exclude_fields=None):
    exclude_fields = exclude_fields or []
    input_string = "MTG Card: "

    for field, value in example.items():
        # Skip fields in exclude_fields
        if field in exclude_fields:
            continue

        # Check for None, NaN, empty list, empty dict, or empty string and add <no_fieldX> if missing
        if (value is None or
            (isinstance(value, float) and math.isnan(value)) or
            (isinstance(value, list) and len(value) == 0) or
            (isinstance(value, str) and value == "") or
            (isinstance(value, dict) and not value)):  # Check if dict is empty
            input_string += f"<{field}> <no_{field}> "
        else:
            # If the field is a list, join its elements as a space-separated string
            if isinstance(value, list):
                value = " ".join(map(str, value))
            # If the field is a dictionary, convert it into a string of key-value pairs
            elif isinstance(value, dict):
                # Convert the dictionary to a string of key:value pairs, ensuring all values are strings
                value = " ".join([f"{k}:{str(v)}" for k, v in value.items()])
            input_string += f"<{field}> {value} "

    example["input"] = input_string.strip()  # Strip extra whitespace at the end
    return example

# Specify fields to exclude
exclude_fields = ["id", "rulings_uri", "input"]

# Apply the function to add the "input" column
dataset = dataset.map(lambda x: create_input(x, exclude_fields))

# Display the updated dataset
print(dataset)
```

```python
from datasets import load_from_disk

# Load the JSON file in streaming mode
dataset = load_from_disk("/content/drive/MyDrive/datasets/MTG/scryfall_final/all_cards")
```

```python
dataset.to_pandas()
```

```python
from transformers import AutoModelForCausalLM

model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-3.2-1B-Instruct", num_labels=5)
```

```python
model.config.max_position_embeddings
```

```python
from transformers import DataCollatorForLanguageModeling, DataCollatorWithPadding
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-1B-Instruct")
tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.pad_token_id

max_length = 5020

data_collator = DataCollatorWithPadding(tokenizer)
```

```python
print(len(tokenizer))  # Check how many tokens are in the tokenizer's vocabulary
```

```python
def tokenize_function(examples):
    return tokenizer(examples["input"], max_length=max_length, padding='max_length', truncation=True)


tokenized_datasets = dataset.map(tokenize_function, batched=True)

#tokenized_inputs = tokenizer(examples["input"], truncation=True, max_length=2048)  # or any other appropriate value
```

```python
df = tokenized_datasets.to_pandas()
df
```

```python
filtered_df = df[df['input_ids'].apply(lambda lst: len(lst) > 5020)]
filtered_df
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(output_dir="test_trainer", eval_strategy="epoch")
```

```python
import numpy as np
import evaluate

metric = evaluate.load("accuracy")
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets,
    tokenizer=tokenizer
)
```

```python
trainer.train()
```
